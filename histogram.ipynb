```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Baca citra dan hitung histogram
img = cv2.imread('cat.jpg', 0)
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
```

```python
# Plot histogram
plt.figure()
plt.title('Histogram Citra Kucing (Grayscale)')
plt.xlabel('Intensitas Pixel')
plt.ylabel('Frekuensi')
plt.plot(hist)
plt.xlim([0, 256])
plt.show()
```

```python
# Statistik
mean = np.mean(img)
var = np.var(img)
std = np.std(img)

print(f'Rata-rata: {mean:.2f}')
print(f'Variansi: {var:.2f}')
print(f'Standar Deviasi: {std:.2f}')
```
